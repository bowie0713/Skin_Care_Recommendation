```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_excel("/Users/bowiechuang/Downloads/Chat_generate_new.xlsx")
df
```

```python
df_recommend = df[['USER', "Product Name", "Product Category", "Product Price", 
                   "Product Rating", "Product Reviews", ""]]
```

```python
df_recommend2 = df[["Product Name", "Skin Type"]]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
skintype_combined = df_recommend2['Skin Type'].str.replace(',', ' ')
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(skintype_combined)

cosine_similarity = cosine_similarity(tfidf_matrix)

similarity_df = pd.DataFrame(cosine_similarity, index=df_recommend2['Product Name'], columns=df_recommend2['Product Name'])
```
